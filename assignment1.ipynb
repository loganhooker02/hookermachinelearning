```python
##Import Libraries
from re import M
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#PROBLEM 1
#Extract Data and set up variables
m = 100
data = pd.read_csv('D3.csv')
X1t = data.loc[:,'X1']
X1t = X1t.values.reshape(m, 1)

X2t = data.loc[:,'X2']
X2t = X2t.values.reshape(m, 1)

X3t = data.loc[:,'X3']
X3t = X3t.values.reshape(m, 1)
X_0 = np.ones((m, 1))
X1 = np.hstack((X_0, X1t))
X2 = np.hstack((X_0, X2t))
X3 = np.hstack((X_0, X3t))

Y = data.loc[:,'Y']
theta0 = np.zeros(2, )
```

```python
def computeCost(X,y,theta):
  m = len(X)
  predictions = X.dot(theta)
  errors = np.subtract(predictions, y)
  sqrErrors = np.square(errors)
  J = 1 / (2 * m) * np.sum(sqrErrors)
  return J

def gradientDescent(X, y, theta, alpha, iterations):
  m = len(X)
  cost_history = np.zeros(iterations)
  for i in range(iterations):
    predictions = X.dot(theta)
    errors = np.subtract(predictions, y)
    sum_delta = (alpha / m) * X.transpose().dot(errors);
    theta = theta - sum_delta;
    cost_history[i] = computeCost(X, y, theta)
  return theta, cost_history

rate = 0.08
iterations = 1500;
thetaX1, cost_historyX1 = gradientDescent(X1, Y, theta0, rate, iterations);
thetaX2, cost_historyX2 = gradientDescent(X2, Y, theta0, rate, iterations);
thetaX3, cost_historyX3 = gradientDescent(X3, Y, theta0, rate, iterations);

print('Final value of theta for X1 =', thetaX1)
print('Final value of theta for X2 =', thetaX2)
print('Final value of theta for X3 =', thetaX3)
```

```python
##Linear Regression for X1
print('Final value of theta for X1 =', thetaX1)
print('cost_history for X1 =', cost_historyX1)
plt.figure(figsize=(10, 6))
plt.scatter(X1[:,1], Y, color='red', marker= '+', label= 'Training Data')
plt.plot(X1[:,1],X1.dot(thetaX1), color='green', label='Linear Regression')
plt.grid()
plt.xlabel('X1')
plt.ylabel('Y')
plt.title('X1 Linear Regression Fit')
plt.legend()
plt.figure(figsize=(10, 6))
plt.plot(range(1, iterations + 1),cost_historyX1, color='blue')
plt.rcParams["figure.figsize"] = (10,6)
plt.grid()
plt.xlabel('Number of iterations')
plt.ylabel('Cost (J)')
plt.title('Convergence of gradient descent for X1')
```

```python
##Linear Regression for X2
print('Final value of theta for X2 =', thetaX2)
print('cost_history for X2 =', cost_historyX2)
plt.figure(figsize=(10, 6))
plt.scatter(X2[:,1], Y, color='red', marker= '+', label= 'Training Data')
plt.plot(X2[:,1],X2.dot(thetaX2), color='green', label='Linear Regression')
plt.grid()
plt.xlabel('X2')
plt.ylabel('Y')
plt.title('X2Linear Regression Fit')
plt.legend()
plt.figure(figsize=(10, 6))
plt.plot(range(1, iterations + 1),cost_historyX2, color='blue')
plt.rcParams["figure.figsize"] = (10,6)
plt.grid()
plt.xlabel('Number of iterations')
plt.ylabel('Cost (J)')
plt.title('Convergence of gradient descent for X2')
```

```python
##Linear Regression for X3
print('Final value of theta for X3 =', thetaX3)
print('cost_history for X3 =', cost_historyX3)
plt.figure(figsize=(10, 6))
plt.scatter(X3[:,1], Y, color='red', marker= '+', label= 'Training Data')
plt.plot(X3[:,1],X3.dot(thetaX3), color='green', label='Linear Regression')
plt.rcParams["figure.figsize"] = (10,6)
plt.grid()
plt.xlabel('X3')
plt.ylabel('Y')
plt.title('X3 Linear Regression Fit')
plt.legend()
plt.figure(figsize=(10, 6))
plt.plot(range(1, iterations + 1),cost_historyX3, color='blue')
plt.rcParams["figure.figsize"] = (10,6)
plt.grid()
plt.xlabel('Number of iterations')
plt.ylabel('Cost (J)')
plt.title('Convergence of gradient descent for X3')
```
